```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data into a Pandas DataFrame
data = pd.read_csv(r"D:\Road accident.csv")
# Exploratory Data Analysis (EDA)
```

```python
data
```

```python
data.isnull().sum()
```

```python
# Visualize accident frequency by age_band_of_driver
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='Age_band_of_driver')
plt.title('Accident Frequency by Age Band of Driver')
plt.show()
```

```python
# Visualize accident frequency by road_surface_type
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='Road_surface_type')
plt.title('Accident Frequency by Road Surface Type')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualize accident frequency by light_conditions
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='Light_conditions')
plt.title('Accident Frequency by Light Conditions')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualize accident frequency by weather_conditions
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='Weather_conditions')
plt.title('Accident Frequency by Weather Conditions')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualize the gender distribution of drivers involved in accidents
plt.figure(figsize=(6, 6))
data['Sex_of_driver'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Gender Distribution of Drivers')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Visualize the distribution of driver ages
plt.figure(figsize=(8, 5))
sns.boxplot(data=data, x='Age_band_of_driver', y='Accident_severity')
plt.title('Age Distribution vs. Accident Severity')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a pair plot to visualize relationships between numerical variables
sns.pairplot(data=data, vars=['Age_band_of_driver', 'Accident_severity'])
plt.suptitle('Pair Plot for Relationships')
plt.show()
```
